```python
# Import standard python libraries
import os
import pandas as pd
import numpy as np
# Import plotting functions
import matplotlib.pyplot as plt
import seaborn as sb
# Import file readers
import csv
# Import built-in sk machine learning models
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
# Import data analyzation metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# Import image displays
from IPython.display import Image
from IPython.display import display
```

```python
# Import a working path directory
directory = os.getcwd()
# Connect with input directory
input_directory = os.path.join(directory, 'input')
# Connect names to a data frame
names_path = os.path.join(input_directory, 'spambasenames.csv')

# Get the names of features from the csv file
features = pd.read_csv(names_path)
feature_names = features.columns
spam_names = features.copy() 

# Append an extra column to keep the target feature in
target_label = 'Label'
spam_names[target_label] = []
# Save the names as a list
col_names = spam_names.columns
# Create an email data frame with the given name columns
data_path = os.path.join(input_directory, 'spambase.csv')
# Create a data frame w/ spam base data set with the col names
email_data = pd.read_csv(data_path, header=None, names=col_names, index_col=False)
```

```python
# Q1: Simple counts of spam and ham messages.

# Count number of ham and spam messages
counts = email_data[target_label].value_counts()
ham_count = counts[0]
spam_count = counts[1]
# Keep track of rows and cols in total data set
rows = email_data.shape[0]
cols = email_data.shape[1]
# Print the result
print("QUESTION 1")
print("Ham Count:", ham_count)
print("Spam Count:", spam_count)
```

```python
# Q2. Part 1: Does the length of messages indicate spam/ham separation? 

# Create sperate data frames for spam and ham
ham_set = email_data.loc[email_data[target_label] == 0]
spam_set = email_data.loc[email_data[target_label] == 1]
# Drop target feature
non_words = ['capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total', target_label]
ham_row = ham_set.copy().drop(non_words, axis = 1)
spam_row = spam_set.copy().drop(non_words, axis = 1)

# Find the sum of each row
ham_row_sum = ham_row.sum(axis=1)
spam_row_sum = spam_row.sum(axis=1)
# Find the total number of words per email
total_spam_words = spam_row_sum.sum() / spam_count
total_ham_words = ham_row_sum.sum() / ham_count

# Calculate the mean of the sums of each row for spam and ham
print("QUESTION 2: PART 1")
print("The average number of words classified for each label-")
print("For Ham: %.2f" %total_ham_words)
print("For Spam: %.2f" %total_spam_words)
print("This calculation would imply that an email containing more words within the dataset is more likely to " 
      "be spam.")
```

```python
# We want to solve for the words with the highest probability to be spam and ham compared to all emails. 
probability_ham = ham_row.sum() - spam_row.sum()
probability_spam = spam_row.sum() - ham_row.sum()
```

```python
# Finds the top 2 words found in an email
def max_words(prob_df):
    # Save the words with the highest mean instance, along with their indexes
    max1 = 0
    max2 = 0
    max_index1 = 0
    max_index2 = 0
    # Iterate through the array to find the 2 largest words in the df
    for i in range(prob_df.shape[0]):
        if (prob_df[i] > max1):
            # Give second largest value to second largest value
            max2 = max1
            max_index2 = max_index1
            # Set the new max value to the current one
            max1 = prob_df[i]
            max_index1 = i
        # If a value is not larger than the first, but is larger than the second then make it second largest
        elif (prob_df[i] > max2):
            max_index2 = i
            max2 = prob_df[i]
    return max1, max2, max_index1, max_index2

# Call the function max_words to solve the 2 most probable words for ham
ham_max1, ham_max2, ham_max_index1, ham_max_index2 = max_words(probability_ham)
ham1 = probability_ham.index.values[ham_max_index1]
ham2 = probability_ham.index.values[ham_max_index2]

# Find the top 2 words which are found in a ham email
spam_max1, spam_max2, spam_max_index1, spam_max_index2 = max_words(probability_spam)
spam1 = probability_spam.index.values[spam_max_index1]
spam2 = probability_spam.index.values[spam_max_index2]

print("Most Common Ham Words: ")
print(ham1)
print(ham2)
print("\nMost Common Spam Words: ")
print(spam1)
print(spam2)
```

```python
# Q2. Part 2: How about the presence or absence of certain words like ‘free’? 

# Print results for the 2 most common spam and ham words and their mean categogircal difference from all emails
print("QUESTION 2: PART 2")
print('Some emails are more inclined to be spam or ham based on the occurrences of the words out of all the '
      'words in the email. Here are the mean percentages of words which show how much more frequently '
      'these words are used on average compared to the opposing email classification.\n')
print("Spam mean of",spam1,": %.1f" %spam_max1)
print("Spam mean of",spam2,": %.1f" %spam_max2)
print("Ham mean of",ham1,": %.1f" %ham_max1)
print("Ham mean of",ham2,": %.1f" %ham_max2)
```

```python
# Q2. Part 3: Plot the distributions and spread of the two classes across a few such features that you construct. 

# Create a data frame which contains our selected features and target
selected_features = email_data[[spam1, spam2, ham1, ham2, target_label]].copy()

# Plot the spread of features as a dot plot first
print("QUESTION 2: PART 3")
print("Spread of Selected Features - May take a moment, please be patient. Thank you!")
spread = sb.pairplot(selected_features, hue=target_label, diag_kws={'bw': 0.2})
```

```python
# Q2. Part 4: Plot the distributions of the two classes across a few such features that you construct. 

# Plot the distribution of features
def plot_histogram(feature, my_df, file_name):
    # Create a distribution graph for "free"
    for classifications in my_df[target_label].unique():
        # Subset to the classification
        subset = my_df[my_df[target_label] == classifications]
        # Draw the density plot
        sb.distplot(subset[feature], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 1.5},
                 label = classifications)
    # Plot formatting
    plt.legend(prop={'size': 10}, title = 'Spam=1, NotSpam=0')
    plt.title('Distribution for Frequency of %s' %feature)
    plt.xlabel('Frequency of %s' %feature)
    plt.ylabel('Density')
    # Connect names to a data frame
    names_path = os.path.join(output_directory, file_name)
    # Save distribution graph to an image
    plt.savefig(names_path)
    # Clear the image
    plt.clf()
    return names_path

# Make a distribution histogram for a ham feature and a spam feature
ham_path = plot_histogram(ham1, selected_features, "ham_feature.png")
spam_path = plot_histogram(spam2, selected_features, "spam_feature.png")

# Get the two images from their output paths and display them
print("QUESTION 2: PART 4\n")
print("Showing the distrbution of a ham word vs a spam word")
histo1 = Image(filename=ham_path) 
histo2 = Image(filename=spam_path) 
display(histo1, histo2)
```

```python
# Shuffle positive and negative data sets
df_shuffled=email_data.iloc[np.random.permutation(email_data.index)].reset_index(drop=True)

# Split dataset into training set and test set.
X_train, X_test, Y_train, Y_test = train_test_split(df_shuffled[feature_names], df_shuffled[target_label], test_size=0.3, random_state=1) # 70% training and 30% test

# Concatenate train and test set back together
train_set = pd.concat([X_train, Y_train], axis=1)
test_set = pd.concat([X_test, Y_test], axis=1)
```

```python
# Create an sklearn logistic regression model
logmodel = LogisticRegression(solver='lbfgs', max_iter=3_000)
# Create an sklearn gaussian naive bayes classifier model
gnb = GaussianNB()

# Fit both models to the training set
logmodel = logmodel.fit(X_train, Y_train)
gnb = gnb.fit(X_train, Y_train)
```

```python
# Q3. Part 1: Use sklearn logistic regression and sklearn to compare the results using 10 fold cross validation
# and what is the accuracy of both?

# Calculate the cross validation accuracy
log_score = cross_val_score(logmodel, X=X_train, y=Y_train, cv=10, n_jobs=10)
gnb_score = cross_val_score(gnb, X=X_train, y=Y_train, cv=10, n_jobs=1)

print('QUESTION 3: PART 1\n')
print("10 Fold Cross Validation Accuracy - Logistic Regression: %.3f" %log_score.mean())
print("10 Fold Cross Validation Accuracy - Naive Bayes: %.3f" %gnb_score.mean())
print("The Logistic Regression models seems to be classifying the data more accurately.")
```

```python
# Question 3. Part 2: What is the accuracy of both of the models?

print("QUESTION 3: PART 2")

# Have both models estimate the test set
log_y_pred = logmodel.predict(X_test)
gnb_y_pred = gnb.predict(X_test)
# Calculate the accuracy of both models
log_accuracy = metrics.accuracy_score(Y_test, log_y_pred)
gnb_accuracy = metrics.accuracy_score(Y_test, gnb_y_pred)
print("SK Logistic Regression Accuracy %.3f" %log_accuracy)
print("SK Naive Bayes Accuracy %.3f" %gnb_accuracy)
```

```python
# Q3. Part 3: Plot the two AUCs.

print("QUESTION 3: PART 3")

# Calculate the probabilities for each model
log_prob = logmodel.predict_proba(X_test)[:, 1]
gnb_prob = gnb.predict_proba(X_test)[:, 1]

# Calculate roc curve logistics for both models
logistic_fpr, logistic_tpr, threshold_log = roc_curve(Y_test.values, log_prob)
gnb_fpr, gnb_tpr, threshold_gnb = roc_curve(Y_test.values, gnb_prob)

# Calculate the AUC plot for both models as specificity and 1 - sensitivity
auc_logistic = auc(logistic_fpr, logistic_tpr)
auc_gnb = auc(gnb_fpr, gnb_tpr)
# Plot a figure with the AUC for Naive Bayes and Logistic Regression
plt.figure(figsize=(5, 5), dpi=80)
plt.plot(gnb_fpr, gnb_tpr, linestyle='-', label='Naive Bayes (auc = %0.3f)' % auc_gnb)
plt.plot(logistic_fpr, logistic_tpr, marker='.', label='Logistic Regressions (auc = %0.3f)' % auc_logistic)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')
plt.legend()
plt.show()
```
